```python
import io
from collections import Counter
from konlpy.tag import *
import pandas as pd
```

```python
def get_text(filename):
    dlg_list = []
    with io.open(filename, encoding='euc-kr') as f:
        i=0
        for a in f:
            b = a.strip().split('\t')
            if len(b) > 2 and b[1] != '<AC>':
                dlg_list.append(b[2])
    return dlg_list
```

```python
dlg_list= get_text("Fashion-How/data/ddata.txt")
```

```python
dlg_list = list(map(str.strip, dlg_list))
```

```python
dlg_list[:10]
```

```python
dlg = " ".join(dlg_list)
```

```python
dlgword_list = dlg.split(" ")
```

```python
for idx, word in enumerate(dlgword_list):
    dlgword_list[idx] = "".join(list(map(lambda word : word if word != '.' and word != ',' and word != '!' and word != '?' else '',word)))
```

```python
word_freq = Counter(dlgword_list)
print(word_freq)
```

```python
mecab = Mecab(dicpath=r"C:/mecab/mecab-ko-dic")
```

```python
dlgtok = " ".join(mecab.morphs(" ".join(dlgword_list)))
dlgtok_list = dlgtok.split()
```

```python
c = Counter
tokword_freq = c(dlgtok_list)
```

```python
word_freq = pd.DataFrame(list(tokword_freq.items()),columns=['Word', 'Frequency'])
```
